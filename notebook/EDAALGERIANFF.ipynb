## DATA CLEANING

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df = dataset
```

```python
df.head()
```

```python
df.tail()
```

because the region values are "float" type, the below code converts it into "int"

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#removing null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
 df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
#fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

### Change the required column as integer data type

```python
df.columns
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

### Change the required column to float data type

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects :
    if i !='Classes':   #except classes , all col should be float
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
#saving the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

## Exploratory Data Analysis

```python
#drop day,month,year
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
#categoris in classes
df_copy['Classes'].value_counts()
```

```python
#encoding of categories in classes
df_copy['Classes']= np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
#!pip install seaborn
#!pip install --upgrade matplotlib
```

```python
##plot density plot for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
#percentage for pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
#plotting pie chart
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
